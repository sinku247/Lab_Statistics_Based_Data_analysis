```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#상향 추세 시각화

np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100).cumsum()

#데이터 프레임 생성
data = pd.DataFrame({
    'Date': dates,
    'Values' : values
})

print(data)
```

```python
# 상향 추세 시각화

data['Upward_Trend'] = np.linspace(0, 50, len(data)) #상향 추세 생성
```

```python
plt.plot(data['Date'], data['Upward_Trend'], label='Upward Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Upward Trend')
plt.legend()
plt.show
```

```python
data['Downward_Trend'] = np.linspace(50,0,len(data))
```

```python
plt.plot(data['Date'], data['Downward_Trend'], label='Downward Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Downward Trend')
plt.legend()
plt.show
```

```python
data['Flat_Trend'] = np.full(len(data), 50) # 평평한 추세 생성
```

```python
plt.plot(data['Date'], data['Flat_Trend'], label='Flat Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Flat Trend')
plt.legend()
plt.show
```
