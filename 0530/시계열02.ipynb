#시계열 데이터 탐색 및 시각화

-시계열 데이터 시각화 소개 (선 그래프, 박스 플롯, 자기상관 그래프)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시계열 데이터 생성

np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)


# 데이터 프레임 생성

data = pd.DataFrame({
    'Date': dates, 'Values':values
})

# 시계열 데이터 시각화
plt.plot(data['Date'], data['Values'])
plt.xticks(rotation=45)
plt.show
```

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

# 데이터 프레임 생성

data = pd.DataFrame({
    'Date': dates, 'Values':values
})
```

```python
#시계열 데이터의 막대 그래프
plt.bar(data['Date'], data['Values'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Scatter Plot')
plt.show
```

```python
histogram_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
histogram_val = np.random.randn(1000)

histogram_data = pd.DataFrame({
    'Date' : histogram_date,
    'Value' : histogram_val
})

#print(histogram_data)

plt.hist(histogram_data['Value'], bins=20)
plt.show
```

```python
#히트맵 시각화

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


#시계열 데이터 생성
hit_date = pd.date_range(start='2023-01-01', periods=1000, freq='D')
hit_val = np.random.randn(1000)

hit_data = pd.DataFrame({
    'Date' : hit_date,
    'Value' : hit_val
})
```

```python
# 데이터를 피벗 테이블 변환
pivot_table = hit_data.pivot(index='Date', columns='Value', values='Value')
```

```python
#히트맵 시각화

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, cmap='YlOrRd', cbar=True)
plt.title('Time Series Heatmap')
plt.xlabel('Value')
plt.ylabel('Date')
plt.show()
```
